```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfv = TfidfVectorizer(min_df=3,  max_features=None,
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
print(sig[0])
```

```python
def give_recomendations(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movies_cleaned['original_title'].iloc[movie_indices]
```

# Prepare metadata

```python
import ast

def get_names(text):
    try:
        return " ".join([i['name'] for i in ast.literal_eval(text)])
    except:
        return ""
```

# Extract useful features

```python
def get_top_cast(text, n=3):
    try:
        return " ".join([i['name'] for i in ast.literal_eval(text)[:n]])
    except:
        return ""

def get_director(text):
    try:
        for i in ast.literal_eval(text):
            if i['job'] == 'Director':
                return i['name']
        return ""
    except:
        return ""
```

# Create a tags column

```python
movies_cleaned['genres'] = movies_cleaned['genres'].apply(get_names)
movies_cleaned['keywords'] = movies_cleaned['keywords'].apply(get_names)
movies_cleaned['cast'] = movies_cleaned['cast'].apply(get_top_cast)
movies_cleaned['director'] = movies_cleaned['crew'].apply(get_director)

movies_cleaned['tags'] = (
    movies_cleaned['overview'].fillna('') + ' ' +
    movies_cleaned['genres'] + ' ' +
    movies_cleaned['keywords'] + ' ' +
    movies_cleaned['cast'] + ' ' +
    movies_cleaned['director']
)
```

# Vectorize with TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(
    stop_words='english',
    max_features=15000
)

tfidf_matrix = tfidf.fit_transform(movies_cleaned['tags'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity

cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)
```

```python
indices = pd.Series(movies_cleaned.index, index=movies_cleaned['original_title']).drop_duplicates()

def recommend(movie, n=10):
    if movie not in indices:
        return "Movie not found."

    idx = indices[movie]
    scores = list(enumerate(cosine_sim[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    scores = scores[1:n+1]

    movie_indices = [i[0] for i in scores]
    return movies_cleaned[['original_title','vote_average']].iloc[movie_indices]
```

```python
recommend('Avatar')
```

# Dimensionality Reduction (Truncated SVD)

```python
from sklearn.decomposition import TruncatedSVD

svd = TruncatedSVD(n_components=2, random_state=42)
tfidf_2d = svd.fit_transform(tfidf_matrix)
```

```python
viz_df = pd.DataFrame({
    'x': tfidf_2d[:, 0],
    'y': tfidf_2d[:, 1],
    'title': movies_cleaned['original_title']
})
```

```python
import matplotlib.pyplot as plt

def plot_similarity(movie_title, n=10):
    if movie_title not in indices:
        print("Movie not found")
        return

    idx = indices[movie_title]
    scores = list(enumerate(cosine_sim[idx]))
    scores = sorted(scores, key=lambda x: x[1], reverse=True)
    scores = scores[1:n+1]

    rec_indices = [i[0] for i in scores]

    plt.figure(figsize=(10, 8))

    # All movies
    plt.scatter(viz_df['x'], viz_df['y'], alpha=0.2, label='All Movies')

    # Selected movie
    plt.scatter(
        viz_df.loc[idx, 'x'],
        viz_df.loc[idx, 'y'],
        color='red',
        s=120,
        label=movie_title
    )

    # Recommended movies
    plt.scatter(
        viz_df.loc[rec_indices, 'x'],
        viz_df.loc[rec_indices, 'y'],
        color='blue',
        s=80,
        label='Recommended'
    )

    plt.legend()
    plt.title(f"Visual Similarity for '{movie_title}'")
    plt.xlabel("Component 1")
    plt.ylabel("Component 2")
    plt.show()
```

```python
plot_similarity('Avatar')
```

# Explainable + Popularity-Aware Recommender

```python
def get_shared_keywords(idx1, idx2, top_n=5):
    vec1 = tfidf_matrix[idx1].toarray()[0]
    vec2 = tfidf_matrix[idx2].toarray()[0]

    common_indices = np.argsort(vec1 * vec2)[::-1]
    feature_names = tfidf.get_feature_names_out()

    keywords = []
    for i in common_indices:
        if vec1[i] > 0 and vec2[i] > 0:
            keywords.append(feature_names[i])
        if len(keywords) == top_n:
            break

    return keywords
```

```python
def recommend_plus(movie, n=10):
    if movie not in indices:
        return "Movie not found."

    idx = indices[movie]
    scores = list(enumerate(cosine_sim[idx]))

    results = []
    for i, sim in scores:
        if i != idx:
            popularity_boost = movies_cleaned.iloc[i]['vote_average'] / 10
            final_score = sim * popularity_boost

            results.append((
                i,
                sim,
                final_score
            ))

    # Sort by final score
    results = sorted(results, key=lambda x: x[2], reverse=True)[:n]

    output = []
    for i, sim, final_score in results:
        explanation = get_shared_keywords(idx, i)
        output.append({
            "Movie": movies_cleaned.iloc[i]['original_title'],
            "Similarity": round(sim, 3),
            "Final Score": round(final_score, 3),
            "Why Recommended": ", ".join(explanation)
        })

    return pd.DataFrame(output)
```

```python
recommend_plus("Avatar")
```
